## Objective: Implement activation function from scratch and apply differentiation
- reference: https://www.v7labs.com/blog/neural-networks-activation-functions

```python
import numpy as np


x = np.array([
    [1,1,0],
    [10,2,-2],
    [-100,23,12]
])
```

### Activation functions:
    - ReLU
    - Sigmoid
    - tanh
    - Leaky ReLU
    - ELU

```python
def relu_activation(x):
    return np.maximum(0,x)

def relu_derivative(x):
    return np.where(x > 0 , 1 , 0)

q = relu_activation(x)
q

#relu differentiation: 1 when x >= 0 and 0 when x < 0
```

```python
def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid_activation(x)
    return s * (1-s)

q = sigmoid_activation(x)
q
```
